```python
#Задание 1
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
feature_names = boston["feature_names"]
data = boston["data"]
```

```python
x = pd.DataFrame(data, columns=feature_names)

x.head()
```

```python
target = boston["target"]
y = pd.DataFrame(target, columns=["price"])

y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})

check_test.head()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
#Задание 2
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators = 1000, max_depth = 12, random_state = 42)
```

```python
model.fit(x_train, y_train.values[:, 0])
```

```python
y_pred2 = model.predict(x_test)
```

```python
check_test2 = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred2": y_pred2.flatten(),
})

check_test2.head()
```

```python
r2_score(y_test, y_pred2)
```

```python
# Модель RandomForestRegressor работает лучше
```

```python
#Задание 4
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df['Class'].value_counts(normalize=True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
x = df.drop('Class', axis=1)
```

```python
y = df['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100, stratify=y)
```

```python
parameters = [{'n_estimators': [10, 15], 'max_features': np.arange(3, 5), 'max_depth': np.arange(4, 7)}]
```

```python
my_grid = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
```

```python
my_grid.fit(x_train, y_train)
```

```python
my_grid.best_params_
```

```python
my_grid = RandomForestClassifier(max_depth = 6, max_features = 3, n_estimators = 15)

my_grid.fit(x_train, y_train)
```

```python
y_pred = my_grid.predict_proba(x_test)
```

```python
y_pred_proba = y_pred[:, 1]
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred_proba)
```
